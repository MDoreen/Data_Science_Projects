```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.preprocessing import OneHotEncoder
```

```python
# Load cleaned dataset
# Source > https://www.kaggle.com/datasets/saurabhshahane/road-traffic-accidents
df = pd.read_csv("/content/drive/MyDrive/RoadAccidents/dataset/cleaned.csv")

```

```python
# Define independent and dependent variables
X = df[['Weather_conditions', 'Type_of_collision']]
y = df['Accident_severity']

```

```python
# Use OneHotEncoder to convert categorical variables to numerical values
one_hot_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

```

```python
# Apply one-hot encoding to all categorical columns
X_encoded = one_hot_encoder.fit_transform(X)

```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
# Save the model
with open('linear_regression_model.pkl', 'wb') as f:
    pickle.dump(model, f)

```

```python
# Predict with the model
# Let's use the model to predict the accident severity for a hypothetical set of independent variables.
# For example, let's say the weather condition is 'Rainy' and the type of collision is 'Vehicle with vehicle collision'.
new_data = {
    'Weather_conditions': ['Rainy'],
    'Type_of_collision': ['Vehicle with vehicle collision']
}
new_df = pd.DataFrame(new_data)

```

```python
# Convert the new data to one-hot encoded format
new_df_encoded = one_hot_encoder.transform(new_df)

```

```python
predicted_severity = model.predict(new_df_encoded)
print(f'Predicted Accident Severity: {predicted_severity[0]}')
```
